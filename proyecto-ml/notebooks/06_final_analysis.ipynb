## Conclusiones Finales

- **Calidad y preparación de datos:** Se realizó una limpieza exhaustiva, codificación y generación de nuevas variables (feature engineering), lo que permitió mejorar la representatividad y utilidad del dataset para los modelos.
- **Modelos supervisados:** El modelo de regresión XGBoost mostró un buen poder predictivo (R² > 0.79, MAE < 0.16), mientras que el árbol de decisión para clasificación alcanzó una exactitud superior al 88%, con buen balance entre precisión y recall. El ajuste de hiperparámetros y el manejo de desbalance contribuyeron a la robustez de los resultados.
- **Modelos no supervisados:** DBSCAN identificó clusters muy densos y bien separados, aunque dejó muchos puntos como ruido. KMeans y el clustering jerárquico agruparon todos los datos, pero con menor separación entre clusters. La elección del modelo depende del objetivo: pureza (DBSCAN) vs. cobertura total (KMeans/Jerárquico).
- **Hallazgos clave:** Se identificaron armas y operadores preferidos, mapas con mayor dificultad y patrones de desempeño que pueden guiar estrategias de juego y balance de diseño.
- **Recomendaciones:** 
    - Profundizar en el análisis de outliers y casos de ruido.
    - Realizar análisis temporal y de evolución de desempeño.
    - Integrar feedback de usuarios/jugadores para validar los hallazgos.
